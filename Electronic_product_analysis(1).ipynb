```python
import pandas as pd
import numpy as np
df=pd.read_csv("Sale_study.csv")
df
```

```python
#1.	Find total sales (Total_Amount).
Total_Sales=df["Total_Amount"].sum()
print("Total Sales:",Total_Sales)
```

```python
# 2.	Calculate average order value.
Avg_VAlue=df["Total_Amount"].mean()
print("Average Value:",Avg_VAlue)
```

```python
# 3.	Count total orders per city.
Total_Order_PerCity=df["City"].value_counts()
print("Count total orders per city:",Total_Order_PerCity)
```

```python
# 4.Find highest and lowest order amount.
Highest_Amount=df["Total_Amount"].max()
Lowest_Amount=df["Total_Amount"].min()
print("Highest_Amount:",Highest_Amount)
print("Lowest_Amount:",Lowest_Amount)
```

```python
# 5.Which product is sold most (based on quantity)?
product=df.groupby("Product")["Quantity"].sum().idxmax()
print("Most Sold Product:",product)
```

```python
# 6.Calculate total sales by City.
sales_by_city=df.groupby("City")["Total_Amount"].sum()
sales_by_city
```

```python
# 7.Calculate total sales by Product.
sales_prod = df.groupby("Product")["Total_Amount"].sum()
print(sales_prod)
```

```python
# 8.Find average discount by Category.
Category=df.groupby("Category")["Discount_%"].mean()
Category
```

```python
# 9.Find total revenue by Payment_Method.
r=df.groupby("Payment_Method")['Total_Amount'].sum() 
r
```

```python
# 10.	Which city generated the highest revenue?
highest_revenue=df.groupby('City')['Total_Amount'].sum().idxmax()
highest_revenue
```

```python
# 11.	Show orders where Discount_% > 10.
dis = df[df["Discount_%"] > 10]
dis
```

```python
orders_mumbai = df[df["City"] == "Mumbai"]
orders_mumbai
```

```python
# 12.Find all orders from Mumbai.
Orders_Mumbai=df.loc[df['City']=='Mumbai']
print(Orders_Mumbai)
```

```python
# 13.	Show orders where Quantity ≥ 3.
df[df['Quantity'] >= 3]
```

```python
# 14.Find orders paid using UPI.
upi=df[df['Payment_Method']=='UPI']
print(upi)
```

```python
# 15.Find Electronics category sales only.
electronics_sales = df[df['Category'] == 'Electronics']
(electronics_sales)
```

```python
# 16.Identify top 5 highest revenue orders.
high_order = df.sort_values(by=['Total_Amount'], ascending=False).head()
print(high_order)
```

```python
# 17.Find monthly sales trend.
df["Order_Date"]=pd.to_datetime(df["Order_Date"])
monthly_sales=df.groupby(df["Order_Date"].dt.month)["Total_Amount"].sum()
print(monthly_sales)
```

```python
# 18.Calculate total revenue after discount.
print("Calculate total revenue after discount",df["Total_Amount"].sum())
```

```python
# 19.Which payment method is most used?
most_pay = df.groupby('Payment_Method')["Total_Amount"].sum().idxmax()
print(most_pay)
```

```python
# 20.	Find repeat customers (if any).
df["Customer_ID"].value_counts()>1
```

```python
#21.Create Profit Column(20%)
df["20%_Profit"]=df["Total_Amount"]*0.20
df
```

```python
#22.Calculate and display City wise Average Value
Avg = df.groupby('City')["Total_Amount"].mean()
print(Avg)
```

```python
#23.Sort Data In Descending Order By Total_Amount
df.sort_values(by=['Total_Amount'], ascending=False)
```
